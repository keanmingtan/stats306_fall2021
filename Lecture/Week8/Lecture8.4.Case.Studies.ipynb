In the following I will provide a summary of regular expressions that we have learnt so far

####  wildcard: `.`: 
Used inside of a regular expression, the period matches any single character.

```r
str_view_all("else every eele etcetera", "e..e ") 
```

#### Character classes
A "character class" is a special pattern that matches a collection of characters. 

`\d` will match any digit
`\s` will match any space


```r
str_view(c("number1", "two", "3hree"), "\\d")
y = c("spa ce", "hello\tworld", "multi\nline")
writeLines(y)
str_view(y, "\\s")
```

You can form your own character class using square brackets: `[abc]` will match *one of* `a`, `b`, or `c`. 

```r
str_view_all("Here is a Sentence", "[A-Z]\\w{3,}")
```

#### Grouping
Sometims we may find it helpful to group a series of regular expression. This is done by putting a paranthesis between the regular expression.

```r
str_view("This will be grouped", "\\b([a-z]+ ?)+")
```

## Backreferences
Parentheses define groups that can be referred to later in the match as `\1`, `\2` etc. This is called a backreference. For example:

    (.)\1

will match the same character repeated twice in a row:

```r
"eeleeleeeellef"  %>% str_view_all("(.)\\1")

```

## Alternatives
An *alternative* means *match this or that*. Alternative patterns can be matched using the syntax `(this|that)`.

#### Create a table that consists of cards drawn by artist with first name John.   

```r
mtg%>% mutate(AvonCard = str_extract_all(mtg$artist,pattern="John\\s\\w{1,}",simplify=TRUE))%>%
filter(AvonCard!="") %>% select(name,AvonCard,manaCost,type,text,rarity)%>% 
print
```

#### What if there is a typo where the capital letter is actually lower case? Write a regular expression to extract those names as well. 

```r
mtg%>% mutate(AvonCard = str_extract_all(mtg$artist,pattern="(j|J)ohn\\s\\w{1,}",simplify=TRUE))%>%
filter(AvonCard!="") %>% select(name,AvonCard,manaCost,type,text,rarity)%>% 
print
```

Now let's try to replace those lower case john to capital letter John 

```r
mtg%>% mutate(AvonCard = str_extract_all(mtg$artist,pattern="(j|J)ohn\\s\\w{1,}",simplify=TRUE))%>%
filter(AvonCard!="") %>% select(name,AvonCard,manaCost,type,text,rarity)%>% 
mutate(AvonCard = str_replace_all(AvonCard[,1],pattern="^j","J"))%>%
print
```

#### Now let us create a table of cards with artists with last name that starts with an A.   

```r
mtg%>% mutate(AvonCard = str_extract_all(mtg$artist,pattern="\\w{1,}\\sA.*$",simplify=TRUE))%>%
filter(AvonCard!="") %>% select(name,AvonCard,manaCost,type,text,rarity)%>% 
print
```

#### In Magic the Gathering, some cards have very powerful ability that allows us to draw additional cards.  Those abilities are desribed in the text column.  Now, let us try to find every cards that contain the word Draw/draw and cards, and output those cards in a table.   

```r
mtg %>% mutate(draw = str_extract_all(mtg$text,pattern="(\\w{1,}\\s)*.*\\s*(d|D)raw(\\s\\w{1,})*\\scard.*(\\s\\w{1,})*",simplify=TRUE))%>%
filter(draw!="") %>% select(name,text)%>% print(width=100)
```

#### In Magic the Gathering, there are five different elements/mana: W G B R U.  Some spells/Creatures that are more powerful multiple same elements to cast.  Identify cards that need at least three of the same mana to cast.      

```r
mtg %>% mutate(draw = str_extract_all(mtg$manaCost,pattern="(\\{\\d\\})*(\\{[WGBRU]\\})\\2{2,}",simplify=TRUE))%>%
filter(draw!="") %>% select(name,manaCost,text,rarity)%>% print(width=100)

```
